```python

import re
import os
from collections import defaultdict
from pdfminer.high_level import extract_text
from nltk.tokenize import RegexpTokenizer
import collections
import matplotlib.pyplot as plt
import math
from math import sqrt
amharic_punc_and_apos = r"[፨፧፦!፥፤፣፣።፡,():;.\'\u1369-\u137C]+"
amharic_punc_and_apos_pattern = re.compile(amharic_punc_and_apos)
amharic_pattern = re.compile(r"[\u1200-\u137F]+")


def Normalize_tokenize(text):
    clean_text = amharic_punc_and_apos_pattern.sub(' ', text)
    tokenized = amharic_pattern.findall(clean_text)
    return tokenized
def remove_index_terms(tokens, percentage_to_remove=0.2):
    word_counts = collections.Counter(tokens)
    total_words = len(tokens)
    sorted_counts = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)
    
    num_index_terms_to_remove = int(percentage_to_remove * len(sorted_counts))
    index_terms = [word for word, _ in sorted_counts[:num_index_terms_to_remove]]
    filtered_tokens = [token for token in tokens if token not in index_terms]
    return filtered_tokens

def analyze_word_frequencies(tokens):
    word_counts = collections.Counter(tokens)
    total_words = len(tokens)
    sorted_counts = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)
    # print(sorted_counts)
    ranks = [i + 1 for i, _ in enumerate(sorted_counts)]
    rank_freq_products = [rank * freq for rank, freq in zip(ranks, [count for _, count in sorted_counts])]

    print("Word Frequencies (Ranked):")
    for rank, (word, count) in enumerate(sorted_counts, start=1):
        relative_freq = (count / total_words) * 100
        print(f"{rank}. {word}: {count} ({relative_freq:.2f}%)")

    plt.figure(figsize=(8, 6))
    plt.scatter(ranks, [count for _, count in sorted_counts], marker='o', color='b')
    plt.xlabel("Rank")
    plt.ylabel("Frequency")
    plt.title("Frequency vs. Rank (Logarithmic Scale)")
    plt.yscale("log")
    plt.grid(True)
    plt.tight_layout()
    plt.show()



# Tokenize and remove index terms
#tokens = Normalize_tokenize(text)


document_id_counter = 0
inverted_index = defaultdict(list)
for filename in os.listdir('C:/Users/Dell/Desktop/pyback'):
  if filename.endswith(".pdf"):
    document_id = document_id_counter  # Assuming document_id_counter is defined elsewhere
    document_id_counter += 1
    text = extract_text(os.path.join('C:/Users/Dell/Desktop/pyback', filename))
tokens = Normalize_tokenize(text)
analyze_word_frequencies(tokens)
tokens_after_removal=remove_index_terms(tokens) #words used for indexing
with open("tokenized.txt", "w", encoding="utf-8") as tobenormalized:
    for words in tokens_after_removal:
        tobenormalized.write(words+" ")
print(tokens_after_removal)
#run time == 

```
