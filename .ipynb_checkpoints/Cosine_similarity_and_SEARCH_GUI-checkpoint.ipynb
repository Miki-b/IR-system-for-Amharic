```python
import re
import os
import pickle
import collections
import math
import PyPDF2
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter as tk
from tkinter import font as tkfont, messagebox, simpledialog
from collections import defaultdict

# Amharic text processing patterns
amharic_punc_and_apos = r"[፨፧፦!፥፤፣።፡,():;.\'\u1369-\u137C]+"
amharic_punc_and_apos_pattern = re.compile(amharic_punc_and_apos)
amharic_pattern = re.compile(r"[\u1200-\u137F]+")

# Function to normalize and tokenize Amharic text
def Normalize_tokenize(text):
    clean_text = amharic_punc_and_apos_pattern.sub(' ', text)
    tokenized = amharic_pattern.findall(clean_text)
    newtext = []
    for words in tokenized:
        words = re.sub('[ሃኀኃሐሓኻ]', 'ሀ', words)
        words = re.sub('[ሑኁዅ]', 'ሁ', words)
        words = re.sub('[ኂሒኺ]', 'ሂ', words)
        words = re.sub('[ኌሔዄ]', 'ሄ', words)
        words = re.sub('[ሕኅ]', 'ህ', words)
        words = re.sub('[ኆሖኾ]', 'ሆ', words)
        words = re.sub('[ሠ]', 'ሰ', words)
        words = re.sub('[ሡ]', 'ሱ', words)
        words = re.sub('[ሢ]', 'ሲ', words)
        words = re.sub('[ሣ]', 'ሳ', words)
        words = re.sub('[ሤ]', 'ሴ', words)
        words = re.sub('[ሥ]', 'ስ', words)
        words = re.sub('[ሦ]', 'ሶ', words)
        words = re.sub('[ዓኣዐ]', 'አ', words)
        words = re.sub('[ዑ]', 'ኡ', words)
        words = re.sub('[ዒ]', 'ኢ', words)
        words = re.sub('[ዔ]', 'ኤ', words)
        words = re.sub('[ዕ]', 'እ', words)
        words = re.sub('[ዖ]', 'ኦ', words)
        words = re.sub('[ጸ]', 'ፀ', words)
        words = re.sub('[ጹ]', 'ፁ', words)
        words = re.sub('[ጺ]', 'ፂ', words)
        words = re.sub('[ጻ]', 'ፃ', words)
        words = re.sub('[ጼ]', 'ፄ', words)
        words = re.sub('[ጽ]', 'ፅ', words)
        words = re.sub('[ጾ]', 'ፆ', words)
        words = re.sub('(ሉ[ዋአ])', 'ሏ', words)
        words = re.sub('(ሙ[ዋአ])', 'ሟ', words)
        words = re.sub('(ቱ[ዋአ])', 'ቷ', words)
        words = re.sub('(ሩ[ዋአ])', 'ሯ', words)
        words = re.sub('(ሱ[ዋአ])', 'ሷ', words)
        words = re.sub('(ሹ[ዋአ])', 'ሿ', words)
        words = re.sub('(ቁ[ዋአ])', 'ቋ', words)
        words = re.sub('(ቡ[ዋአ])', 'ቧ', words)
        words = re.sub('(ቹ[ዋአ])', 'ቿ', words)
        words = re.sub('(ሁ[ዋአ])', 'ኋ', words)
        words = re.sub('(ኑ[ዋአ])', 'ኗ', words)
        words = re.sub('(ኙ[ዋአ])', 'ኟ', words)
        words = re.sub('(ኩ[ዋአ])', 'ኳ', words)
        words = re.sub('(ዙ[ዋአ])', 'ዟ', words)
        words = re.sub('(ጉ[ዋአ])', 'ጓ', words)
        words = re.sub('(ደ[ዋአ])', 'ዷ', words)
        words = re.sub('(ጡ[ዋአ])', 'ጧ', words)
        words = re.sub('(ጩ[ዋአ])', 'ጯ', words)
        words = re.sub('(ጹ[ዋአ])', 'ጿ', words)
        words = re.sub('(ፉ[ዋአ])', 'ፏ', words)
        words = re.sub('[ቊ]', 'ቁ', words)
        words = re.sub('[ኵ]', 'ኩ', words)
        newtext.append(words)
    return newtext

# Functions for TF-IDF and Cosine Similarity
def word_frequency(token, tokens):
    return tokens.count(token)

def calculate_tf(word, tokens):
    word_freq = word_frequency(word, tokens)
    total_words = len(tokens)
    return word_freq / total_words

def calculate_idf(word, all_document_tokens):
    num_documents_with_word = sum(1 for tokens in all_document_tokens if word in tokens)
    total_documents = len(all_document_tokens)
    return math.log(total_documents / (1 + num_documents_with_word), 2)

def calculate_tfidf(term, document_tokens, all_document_tokens):
    tf = calculate_tf(term, document_tokens)
    idf = calculate_idf(term, all_document_tokens)
    return tf * idf

vector = defaultdict(list)
def create_vector(token, tokens, document_id):
    vector[token] = {document_id: word_frequency(token, tokens)}
    return vector

def calculate_cosine_similarity(document_vector, query_vector):
    dot_product = 0.0
    document_vector_magnitude = 0.0
    query_vector_magnitude = 0.0
    
    for term in document_vector:
        if term in query_vector:
            arr1 = list(document_vector[term].values())
            arr2 = list(query_vector[term].values())
            dot_product += arr1[0] * arr2[0]
            document_vector_magnitude += arr1[0]**2
            query_vector_magnitude += arr2[0]**2
    
    document_vector_magnitude = math.sqrt(document_vector_magnitude)
    query_vector_magnitude = math.sqrt(query_vector_magnitude)
    
    if document_vector_magnitude * query_vector_magnitude == 0:
        return 0.0
    
    cosine_similarity = dot_product / (document_vector_magnitude * query_vector_magnitude)
    return cosine_similarity

def rank_documents(document_vectors, query_vector):
    cosine_similarities = []
    
    for document_vector in document_vectors:
        cosine_similarity = calculate_cosine_similarity(document_vector, query_vector)
        doc_id = list(document_vectors[document_vector].keys())[0]
        cosine_similarities.append((doc_id, cosine_similarity))
    
    cosine_similarities.sort(key=lambda x: x[1], reverse=True)
    return cosine_similarities
import os
import glob

# Define the directory where your PDFs are located
pdf_directory = 'path_to_your_pdf_folder'  # Replace with the actual path to your PDF folder

# Get all PDF files from the directory
pdf_paths = glob.glob(os.path.join(pdf_directory, "*.pdf"))
# GUI Functions
def search_pdfs_for_query(query):
    matching_pdfs = []
    for pdf_path in pdf_paths:
        try:
            with open(pdf_path, "rb") as f:
                reader = PyPDF2.PdfFileReader(f)
                num_pages = reader.getNumPages()
                for page_num in range(num_pages):
                    page = reader.getPage(page_num)
                    text = page.extractText()
                    if query.lower() in text.lower():
                        matching_pdfs.append(pdf_path)
                        break
        except Exception as e:
            print(f"Error reading {pdf_path}: {e}")
    return matching_pdfs
search_history = []
def search():
    result_text.delete(1.0, tk.END)
    query = query_entry.get()
    if query:
        processed_query = Normalize_tokenize(query)
        query_vector = defaultdict(list)
        for query_ in processed_query:
            query_vector = create_vector(query_, processed_query, 0)
        for dictionary in data:
            ranked_documents = rank_documents(dictionary, query_vector)
            search_history.append(query)
            for doc_id, cosine_similarity in ranked_documents[:5]:
                result_text.insert(tk.END, f"Document: {doc_id}, Cosine Similarity: {cosine_similarity}\n")
        matching_pdfs = search_pdfs_for_query(query)
        for pdf_path in matching_pdfs:
            result_text.insert(tk.END, f"Found in: {pdf_path}\n")

def reset():
    query_entry.delete(0, tk.END)
    result_text.delete(1.0, tk.END)

def show_history():
    history_window = tk.Toplevel(root)
    history_window.title("Search History")
    history_window.configure(bg='#D0F0C0')

    history_label = tk.Label(history_window, text="Search History:", bg='#D0F0C0', font=font_style)
    history_label.pack(pady=10)

    history_listbox = tk.Listbox(history_window, width=60, height=10, bg='#FFFFFF', font=font_style)
    history_listbox.pack(padx=20, pady=20)

    for query in search_history:
        history_listbox.insert(tk.END, query)

    close_button = tk.Button(history_window, text="Close", command=history_window.destroy, bg='#008CBA', fg='white', font=font_style)
    close_button.pack(pady=10)

    history_window.mainloop()

# Load vector data from file
try:
    with open('data.pickle', 'rb') as f:
        data = pickle.load(f)
except FileNotFoundError:
    print("Vector data file not found.")
    data = []

# Main GUI
root = tk.Tk()
root.title("Amharic Search Engine")
root.configure(bg='#D0F0C0')

font_style = tkfont.Font(family="Verdana", size=12)
query_label = tk.Label(root, text="Enter your search query:", bg='#D0F0C0', font=font_style)
query_label.pack(pady=10)

query_entry = tk.Entry(root, width=50, font=font_style)
query_entry.pack(pady=10)

search_button = tk.Button(root, text="Search", command=search, bg='#4CAF50', fg='white', font=font_style)
search_button.pack(pady=10)

reset_button = tk.Button(root, text="Reset", command=reset, bg='#f44336', fg='white', font=font_style)
reset_button.pack(pady=10)

history_button = tk.Button(root, text="Search History", command=show_history, bg='#008CBA', fg='white', font=font_style)
history_button.pack(pady=10)

result_text = tk.Text(root, width=80, height=20, wrap=tk.WORD, font=font_style)
result_text.pack(pady=20)

root.mainloop()
```
